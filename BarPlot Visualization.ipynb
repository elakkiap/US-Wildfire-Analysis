```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cnx = sqlite3.connect('/Users/jessiarcher/downloads/FPA_FOD_20170508.sqlite')
```

```python
df = pd.read_sql_query("SELECT FIRE_YEAR,STAT_CAUSE_DESCR,LATITUDE,LONGITUDE,STATE,FIRE_SIZE,FIRE_SIZE_CLASS FROM 'Fires'", cnx)
print(df.head())
```

```python
freq_df = df.groupby('FIRE_YEAR').agg('count')
```

```python
freq_df['STATE'].plot(xlabel = 'Year', ylabel = 'Frequency', title = 'Frequency of Fires 1992 to 2015', kind="bar", color='orange')
```

```python
df['STAT_CAUSE_DESCR'].value_counts().plot(xlabel = 'Causes', title = 'Causes of Fires 1992 to 2015', kind='barh',color='orange')
plt.show()
```

```python
df['STATE'].value_counts()[:10].plot(xlabel = 'States', title = 'Top 10 States with Highest frequency of Fires 1992 to 2015', kind='bar',color='orange')
plt.show()
```

```python
size = df['FIRE_SIZE_CLASS'].replace(['A','B','C','D','E','F','G'], ['0-0.25','0.26-9.9','10.0-99.9','100-299','300-999','1000-4999','5000+'])
size.value_counts().plot(xlabel = 'Size(acres)', title = 'Sizes of Fires 1992 to 2015', kind='barh',color='orange')
plt.show()
```
